```python
# Start with loading all necessary libraries
import numpy as np
import pandas as pd
from os import path
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Load in the dataframe
df = pd.read_csv('winemag-data-130k-v2.csv', index_col=0)
# from: https://www.kaggle.com/zynicide/wine-reviews/data
```

```python
# See first 5 rows
df.head()
```

```python
# How many observations and columns?
df.shape
```

```python
print("There are {} observations and {} columns in this dataset.".format(df.shape[0], df.shape[1]))
```

```python
# How many types of wine?
len(df.variety.unique())
```

```python
len(df.country.unique())  # unique values in country
```

```python
# Summary statistics of all countries' points and price
ct = df.groupby("country")
```

```python
ct.describe().head()  # summary of quantitative variables for the first 5 countries
```

```python
# cuntries with the top 5 points
ct.mean().sort_values(by='points', ascending=False).head()
```

```python
# Number of wines by country
plt.figure(figsize=(15,10))
ct.size().sort_values(ascending=False).plot.bar()
plt.xticks(rotation=50)
plt.xlabel("Country of Origin")
plt.ylabel("Number of Wines")
plt.show()
```

```python
?WordCloud
```

```python
text = df.description[0]
```

```python
wordcloud = WordCloud().generate(text)
```

```python
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
wordcloud = WordCloud(max_font_size=50, max_words=100, background_color="white").generate(text)
plt.imshow(wordcloud, interpolation="bilinear")  # make the image appear more smoothly
plt.axis("off")
plt.show()
```

```python
wordcloud.to_file('firstReview.png')
```

```python
# Join all wine reviews into one big text
text = " ".join(review for review in df.description)
```

```python
len(text)
```

```python
useless = set(STOPWORDS)  # copy STOPWORDS into useless
useless.update(["drink", "now", "wine", "flavor", "flavors"])
```

```python
wordcloud = WordCloud(stopwords=useless, background_color="white").generate(text)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")  
plt.axis("off")
plt.show()
```
